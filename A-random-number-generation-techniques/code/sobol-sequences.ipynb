```python
from scipy.stats.qmc import Sobol

# Generate a Sobol sequence of 100 points in 2D
sobol = Sobol(d=2, scramble=False)
quasi_random_points = sobol.random(n=100)
print(quasi_random_points[:5])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random points for comparison
random_points = np.random.rand(100, 2)

# Plot comparison
plt.scatter(
    random_points[:, 0], random_points[:, 1], alpha=0.3, color="red", label="Random"
)
plt.scatter(
    quasi_random_points[:, 0],
    quasi_random_points[:, 1],
    alpha=0.7,
    color="blue",
    label="Sobol",
)
plt.legend()
plt.title("Sobol Sequence vs. Random Sampling")
plt.show()
```

```python
sobol_scrambled = Sobol(d=2, scramble=True)
scrambled_points = sobol_scrambled.random(n=100)
print(scrambled_points[:5])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random points for comparison
random_points = np.random.rand(100, 2)

# Plot comparison
plt.scatter(
    random_points[:, 0], random_points[:, 1], alpha=0.3, color="red", label="Random"
)
plt.scatter(
    quasi_random_points[:, 0],
    quasi_random_points[:, 1],
    alpha=0.5,
    color="blue",
    label="Sobol",
)
plt.scatter(
    scrambled_points[:, 0],
    scrambled_points[:, 1],
    alpha=0.7,
    color="green",
    label="Scrambled Sobol",
)
plt.legend()
plt.title("Sobol Sequence vs. Random Sampling")
plt.show()
```
